#Visualización de datos en Python

Este cuaderno muestra visualizaciones de datos de Python en el conjunto de datos de Iris con las librerías de Seaborn y Pandas.

-Cuadernillo guía para visualizar datos de un csv

```python
#Importar librerias

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
#Extraer conjunto de datos

df_iris = pd.read_csv("/content/Iris.csv")
```

```python
# Ver información general del conjunto de datos. (metadatos)

df_iris.info()
```

```python
#Estadísticas descriptivas

df_iris.describe()
```

```python
#Ver encabezados y final de la tabla

df_iris.head()

df_iris.tail()
```

```python
#Viendo valores faltantes o nulos

df_iris.isnull().sum()

```

```python
#Contar valores de variables categóricas

df_iris["Species"].value_counts()
```

# ¡Grafiquemos!

```python
#Visualización de Distribuciones

df_iris["Species"].hist()

sns.histplot(df_iris["Species"])



# Histograma de cada característica

df_iris.hist(figsize=(10, 8), bins=15)
```

```python
#Boxplot de características para detectar outliers

sns.boxplot(data=df_iris.drop('Id', axis=1), orient="h")
plt.savefig("boxplot_iris.png") #Así guardamos en formato .png las gráficas que vamos generando
df_iris.head()
```

```python
#Visualización de relaciones entre variables

#Diagrama de dispersión o scatter

sns.relplot(data=df_iris, x="SepalLengthCm", y="SepalWidthCm")

# df_iris.plot(kind="scatter",  x="SepalLengthCm", y="SepalWidthCm")

```

```python
#Relación entre variables

sns.pairplot(df_iris, hue="Species", markers=["o", "s", "D"])
```

```python
# Matriz de correlación

# Seleccionar solo las columnas numéricas para la matriz de correlación
df_numerico = df_iris.select_dtypes(include=['float64', 'int64'])

df_numerico.to_csv("iris_numerico.csv", index=False) #Con esto guardamos en un csv nuevo, el dataset manipulado


# Crear la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(df_numerico.corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de correlación")
plt.show()
```

```python
#Grafica de violín

sns.violinplot(x="Species", y="PetalLengthCm", data=df_iris)
```
